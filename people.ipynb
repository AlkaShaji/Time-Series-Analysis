```python
df = df.drop(columns=["Index"])
```

```python
df = df.rename(columns={"User Id": "User_ID",
                        "First Name": "First_Name",
                        "Last Name": "Last_Name",
                        "Sex": "Sex",
                        "Email": "Email",
                        "Phone": "Phone",
                        "Date of birth": "DOB",
                        "Job Title": "Job_Title"
})
```

```python
df.head()
```

```python
new_order = [
    "User_ID", "First_Name", "Last_Name", "Sex", "DOB", "Email", "Phone", "Job_Title"
]
```

```python
df = df.reindex(columns=new_order)
```

```python
df.head()
```

```python
# Count missing values in each column
df.isna().sum()
```

```python
# Or as a percentage
df.isna().mean() * 100
```

```python
# Drop rows with ANY missing value
df.dropna()
```

```python
# Drop rows where ALL values are missing
df.dropna(how='all')
```

```python
# Drop columns with ANY missing value
df.dropna(axis=1)
```

```python
# Fill a single column with a constant
df['Phone'] = df['Phone'].fillna("Not Provided")
```

```python
# Standardize text case
df['First_Name'] = df['First_Name'].str.lower()
```

```python
# Remove leading/trailing spaces
df['Last_Name'] = df['Last_Name'].str.strip()
```

```python
df['Sex'] = df['Sex'].replace({
    "M": "Male",
    "MALE": "Male",
    "male": "Male",
    "F": "Female",
    "FEMALE": "Female",
    "female": "Female"
})
```
