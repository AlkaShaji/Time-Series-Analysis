```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "Boat_id" : [1,2,3,4,5,6],
    "Model": ["Aqualite", "Aualite", "SeaVolt", "SeaVolt", "EcoWave", "EcoWave"],
    "Energy": [120, 115, 140, None, 130, 128],
    "Distance": [30, 32, 40, 38, 42, 40]
}
```

```python
df= pd.DataFrame(data)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
df['Energy'] = df['Energy'].fillna(df['Energy'].mean())
```

```python
df.head()
```

```python
print(df.groupby('Model')['Energy'].mean())
```

```python
df['Efficiency'] = df['Distance'] / df['Energy']
df.head()
```

```python
print(df[['Model', 'Efficiency']].head())
```

```python
speeds = np.array([20, 25, 30, 28, 27])
print("Average speed:", np.mean(speeds))
print("max speed:", np.max(speeds))
```

```python
plt.plot(df['Distance'], df['Energy'], marker='*')
plt.plot("Distance Vs Energy")
plt.xlabel("Distance")
plt.ylabel("Energy")
plt.show()
```
